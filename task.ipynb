```
num_features = X.shape[1]
```

Создаем нейрон с помощью torch.nn.Sequential - контейнер, в котором задаются методы в том порядке, в котором должны будут выполняться.

###Теперь приступим к обучению сети!

Каждый из входных признаков обернем в тензоры (FloatTensor), а тензоры в класс autograd.Variable, который позволит при вызове функции .backward() автоматически вычислять градиент тензора.

```
X = torch.FloatTensor(X)
Y = torch.LongTensor(Y)
```

Код ниже - основной, с которым придется работать при выполнении второго самостоятельного задания.


```
losses = []

# функция потерь
# loss_func = torch.nn.MSELoss(size_average=False)  # квадратичная
# loss_func = torch.nn.CrossEntropyLoss(size_average=False)  # перекрестная энтропия: NLLLoss + LogLoss (для нее шаг побольше)
loss_func = torch.nn.NLLLoss(size_average=False)  # NLLLoss

# скорость обучения (шаг метода оптимизации)
learning_rate = 1e-1

# метод оптимизации
# optimizer = torch.optim.SGD(neuron.parameters(), lr=learning_rate)  # SGD - стохастический градиентный спуск
optimizer = torch.optim.Adam(neuron.parameters(), lr=learning_rate)  # Метод стохастической оптимизации

# количество итераций
num_epochs = 1000

# обучение
for i in range(num_epochs):
    y_pred = neuron(X)

    # Вычисление потерь
    y_pred = torch.stack((torch.FloatTensor([1]) - y_pred, y_pred), dim=1)  # Для перекрестной энтропии и NLL
    loss = loss_func(y_pred, Y)
    losses.append(loss)
    print(f'Iter: {i} -> Loss: {loss.data}')

    # обнуление градиентов, иначе будут копиться градиенты с предыдущей итерации
    optimizer.zero_grad()

    # вычисление градиентов
    loss.backward()

    # обновление весов
    optimizer.step()
```

Тестируем:

```
proba_pred = neuron(X)
y_pred = proba_pred > 0.5
y_pred = y_pred.data.numpy().reshape(-1)

plt.figure(figsize=(10, 8))
plt.scatter(data.iloc[:, 0], data.iloc[:, 1], c=y_pred, cmap='brg')
plt.title('Гавиал и крокодил', fontsize=15)
plt.xlabel('Тон окраски', fontsize=14)
plt.ylabel('Ширина мордочки', fontsize=14)
plt.show();
```

#Самостоятельное задание 3

Аналогично второму заданию необходимо добиться 
лучшего качества нейронной сети, не добавляя дополнительных слоев.

```
import numpy as np
```

```
N = 100
D = 2
K = 3
X = np.zeros((N * K, D))
y = np.zeros(N * K, dtype='uint8')

for j in range(K):
    ix = range(N * j,N * (j + 1))
    r = np.linspace(0.0, 1, N)
    t = np.linspace(j * 4, (j + 1) * 4,N) + np.random.randn(N) * 0.2
    X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
    y[ix] = j

plt.figure(figsize=(10, 10))
plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.brg)
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$y$', fontsize=14)
plt.show();
```

```
X_tensor = torch.FloatTensor(X)
y_tensor = torch.LongTensor(y.astype(np.int64))
```

```
print(X_tensor.shape, y_tensor.shape)
```

### Метод оптимизации Adam

```
losses_Adam = []

ndim_in = 2 # размерность входа (то есть количество признаков)
ndim_out = 3 # размерность выходного слоя (то есть количество классов)
num_epochs = 5000
learning_rate = 1e-1

neuron_Adam = torch.nn.Sequential(
    torch.nn.Linear(ndim_in, ndim_out),
    torch.nn.Sigmoid()
)

loss_fn = torch.nn.CrossEntropyLoss(reduction='sum')

optimizer = torch.optim.Adam(neuron_Adam.parameters(), lr=learning_rate)

for i in range(num_epochs):
    y_pred = neuron_Adam(X_tensor)

    loss = loss_fn(y_pred, y_tensor)
    losses_Adam.append(loss)
    print(f'Iter: {i} -> Loss: {loss.data}')

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()
```

### Метод оптимизации SGD

```
losses_SGD = []

ndim_in = 2 # размерность входа (то есть количество признаков)
ndim_out = 3 # размерность выходного слоя (то есть количество классов)
num_epochs = 5000
learning_rate = 1e-1

neuron_SGD = torch.nn.Sequential(
    torch.nn.Linear(ndim_in, ndim_out),
    torch.nn.Sigmoid()
)

loss_fn = torch.nn.CrossEntropyLoss(reduction='sum')

optimizer = torch.optim.SGD(neuron_SGD.parameters(), lr=learning_rate)

for i in range(num_epochs):
    y_pred = neuron_SGD(X_tensor)

    loss = loss_fn(y_pred, y_tensor)
    losses_SGD.append(loss)
    print(f'Iter: {i} -> Loss: {loss.data}')

    optimizer.zero_grad()

    loss.backward()

    optimizer.step()
```

### График функции потерь

При одинаковых шаге обучения и количестве эпох нейронная сеть с метод оптимизации SGD сначала сходится быстрее на начальных итерациях, но потом сходится быстрее начинает сеть с методом оптимизации Adam

```
plt.figure(figsize=(12, 6))
plt.plot(range(len(losses_Adam)), losses_Adam, c='b')
plt.plot(range(len(losses_SGD)), losses_SGD, c='r')
plt.legend(['Adam', 'SGD'])
plt.grid()
plt.show()
```

Однако использование перекрестной энтропии (Cross Entropy) в качестве функции ошибки дает необычный результат: зеленая ветвь практически игнорируется
